```python
def process_data(file_path):
    with open(file_path, "r") as file:
        content = file.read()
    content = content.lower()
    # words = content.split()
    words = re.findall(r"\b\w+\b", content)
    return words
```

```python
def get_count(corpus):
    words_count = {}
    # words_count = {word: 0 for word in corpus}
    # for word in corpus:
    #     if word in words_count:
    #         words_count[word] += 1
    words_count = Counter(corpus)
    return words_count
```

```python
def get_probs(words_count_dict):
    total_words = sum(words_count_dict.values())
    words_probs = {word: freq / total_words for word, freq in words_count_dict.items()}
    return words_probs
```

```python
def delete_letter(text):
    strings_l = [(text[:i], text[i:]) for i in range(len(text) + 1)]
    deletes_l = [L + R[1:] for L, R in strings_l if L + R[1:] != text and R]
    deletes_s = set(deletes_l)
    return deletes_l
```

```python
def switch_letter(text):
    strings_l = [(text[:i], text[i:]) for i in range(len(text) + 1)]
    switches_l = [L + R[1] + R[0] + R[2:] for L, R in strings_l if len(R) > 1]
    switches_s = set(switches_l)
    return switches_l
```

```python
def replace_letter(text):
    letters = 'abcdefghijklmnopqrstuvwxyz'
    replaces_s = set([text.replace(i, c) for i in text for c in letters])
    replace_l = sorted(list(replaces_s))

    if text in replace_l:
        replace_l.remove(text)
        
    return replace_l 
```

```python
def insert_letter(text):
    letters = 'abcdefghijklmnopqrstuvwxyz'
    strings_l = [(text[:i], text[i:]) for i in range(len(text) + 1)]
    inserts_l = [L + c + R for L, R in strings_l for c in letters]
    return inserts_l
```

```python
def edit_one_letter(word, allow_switches = True):
    delete_l = delete_letter(word)
    inserts_l = insert_letter(word)
    replaces_l = replace_letter(word)
    switches_l = []
    
    if allow_switches:
        switches_l = switch_letter(word)

    edits_l = delete_l + inserts_l + replaces_l + switches_l
    edits_s = set(edits_l)
    return edits_s    
```

```python
edits_s_test = edit_one_letter("at", True)
print(len(edits_s_test))
```

```python
def edit_two_letters(word, allow_switches = True):
    edit_1 = edit_one_letter(word, allow_switches)
    edits_2 = set()
    
    for word_e in edit_1:
        edits_2.update(set(edit_one_letter(word_e, allow_switches)))

    return edits_2
```

```python
def get_corrections(word, vocab, words_prob, n = 2):
    suggestions_l = []
    best_n = []
    temp = []
    
    if word in vocab:
        temp.append(word)

    one_letter_suggestion = sorted(list(edit_one_letter(word, True)))
    for o in one_letter_suggestion:
        if o in vocab:
            temp.append(o)

    two_letter_suggestion = sorted(list(edit_two_letters(word, True)))
    for t in two_letter_suggestion:
        if t in vocab:
            temp.append(t)
    
    temp.append(word)
    
    for i in range(n):
        suggestions_l.append(temp[i])
        best_n.append((temp[i], words_prob[temp[i]]))
    return best_n
```

```python
def min_edit_distance(source, target, ins_cost = 1, del_cost = 1, rep_cost = 2):
    src_letters = ["#"] + list(source)
    target_letters = ["#"] + list(target)

    rows = len(src_letters)
    cols = len(target_letters)
    
    dist_matrix = pd.DataFrame(0, index = src_letters, columns = target_letters)
    
    for i in range(rows):
        dist_matrix.iat[i, 0] = i
    for j in range(cols):
        dist_matrix.iat[0, j] = j
        
    for i in range(1, rows):
        for j in range(1, cols):
            
            if src_letters[i] == target_letters[j]:
                cost = 0
            else:
                cost = rep_cost
                
            dist_matrix.iat[i, j] = min(
                dist_matrix.iat[i - 1, j] + del_cost,
                dist_matrix.iat[i, j - 1] + ins_cost,
                dist_matrix.iat[i - 1, j - 1] + cost
            )

    min_edits = dist_matrix.iat[rows - 1, cols - 1]
    
    return dist_matrix, min_edits
```
